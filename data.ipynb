```python
values = list(map(lambda x: x.split('\t'), text.split('\n')))
values = list(map(lambda x: [x[0], x[1].split(' ~ ')] ,values))
values = list(map(lambda x: [x[0], [x[1][0],x[1][0]] if len(x[1]) == 1 else x[1] ] ,values))
values
```

```python
revalues = []
for row in values:
    tmp = row[0].split(' ')
    if len(tmp) == 1:
        revalues.append( [tmp[0],tmp[0], row[1][0],row[1][1]] )
    else:
        revalues.append( [' '.join(tmp[:-1]),tmp[-1], row[1][0],row[1][1]] )

revalues
```

```python

sql = f'insert into `doso` (`location`, `area`, `start`, `end`) \nvalues \n'

for row in revalues:
    sql += f'(\'{row[0]}\', \'{row[1]}\', \'{row[2]}\', \'{row[3]}\'),\n'
sql = sql[:-2]+';'

print( sql )
```
